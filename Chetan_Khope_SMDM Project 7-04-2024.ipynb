1. If we look at the average 3 month spending of those who have 'rewards' credit card type, spending is not much compared to the other ones such as for e.g. 'chartered' so we should increase the credit card limit on the 'reward' type as this one is the most aquired/used by the users.

3. On the other hand cards such as 'gold' should be offered to those indivisuals who belong to high_salary category from 'B' to 'E' as these indivisuals are not as rich as those who belong to 'A' and 'B', so they may not want highest credit limit but, just more than what is offered by cards such as 'smartearn', 'rewards' which have lower credit limit. Hence 'gold' credit card could be right option for these indivisuals.

4. If we look at the credit cards such as 'aura' 'shoprite' 'cashback' and 'indianoil'  all of them have almost same user numbers, average 3 months of spending, and credit limit. So we are basically offering too many options with no clear or few distinguishing factors and it might be confusing to the customers in terms of which card they should get, so thus reducing the options and focussing on the simplicity yet practicality of cards with clear emphasis on cc_limit, spending and the economical category of the user will lead to the proper offering to the customers based on their preference.
